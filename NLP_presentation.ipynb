```python
tweets = "".join(open('lscp-0.5-fa-normalized.txt').readlines())
```

```python
tweets_num = len(tweets)
for_words_count = len(arman)+len(eng)+len(french)+len(turkish)
fore_list = [arman,eng,french,turkish]
distributions = [[],[],[],[]]

for lang_ind in range(4):
  lang = fore_list[lang_ind]
  for word in lang:
    count = tweets.count(word)
    distributions[lang_ind].append(count)
```

###Mean_usage

```python
mean_dist = [(sum(col)/len(col))/tweets_num for col in distributions]
names = ["Armenain","English","French","Turkish"]
```

```python
for i in range(1,4):
  mean_dist[i]-=(max(distributions[i])/len(distributions[i]))/tweets_num
```

```python
plt.bar(names , mean_dist, 0.8,color = 'grey')
plt.title("Mean usage of words in Persian tweets")
plt.yscale("log")
plt.show()
```

###Top5 Usage

```python
maps = []
for i in range(4):
  res = {fore_list[i][j]: distributions[i][j] for j in range(len(fore_list[i]))}
  maps.append(res)
```

```python
sorted_maps = []
for i in range(4):
  sorted_maps.append(sorted(maps[i].items(),key=lambda x:x[1],reverse=True))
```

```python
import itertools
words_list = []
labels_list = []
for map in sorted_maps:
  for i in range(5):
    words_list.append(map[i][0])
    labels_list.append(map[i][1])
```

```python
top5 = [sum(labels_list[i*5:5*(i+1)]) for i in range(4)]
```

```python
plt.bar(names ,top5 , 0.8,color = 'grey')
plt.title("Count sum of top 5 most used words of each lang in Persian")
plt.yscale("log")
plt.show()
```
